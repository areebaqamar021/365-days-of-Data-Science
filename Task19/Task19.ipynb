# Statistical plotting with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')

tips.replace([np.inf, -np.inf], np.nan, inplace=True)
flights.replace([np.inf, -np.inf], np.nan, inplace=True)
iris.replace([np.inf, -np.inf], np.nan, inplace=True)
```

## Linear Regression Plot 

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Linear Regression Plot: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## Multiple Linear Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, aspect=2, height=6)
plt.title('Multiple Linear Regression Plot: Total Bill vs Tip by Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## Residual Plot

```python
plt.figure(figsize=(10, 6))
sns.residplot(x='total_bill', y='tip', data=tips)
plt.title('Residual Plot: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Residuals')
plt.show()
```

## Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=30, kde=True)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

## Kernel Density Estimate (KDE) Plot

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

## Distribution Plot 

```python
plt.figure(figsize=(10, 6))
sns.displot(tips['total_bill'], bins=30, kde=True)
plt.title('Distribution Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

##  Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## Pair Plot

```python
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

## Heatmap 

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Flight Passengers')
plt.show()
```

## Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', height=8)
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
```

## Facet Grid 

```python
g = sns.FacetGrid(tips, col='smoker', row='sex', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
```
