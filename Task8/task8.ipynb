```python
my_list = [1, 2, 3, 4]

print(my_list[0])  

my_list[1] = 5
print(my_list)  

my_list.append(6)
print(my_list) 

my_list.remove(3)
print(my_list) 

my_tuple = (1, 2, 3, 4)

print(my_tuple[0])  # Output: 1

# my_tuple[1] = 5  # This will raise a TypeError as tuples are immutable

my_set = {1, 2, 3, 4}

my_set.add(5)
print(my_set)

my_set.remove(2)
print(my_set)  

my_set.add(3)
print(my_set)  

my_dict = {'a': 1, 'b': 2, 'c': 3}

print(my_dict['a']) 

my_dict['b'] = 4
print(my_dict)  

my_dict['d'] = 5
print(my_dict) 

del my_dict['c']
print(my_dict)  

# List example
fruits = ['apple', 'banana', 'cherry']
print(fruits)
fruits.append('orange')
print(fruits)  
fruits[1] = 'blueberry'
print(fruits)  
fruits.remove('cherry')
print(fruits)  

# Tuple example
coordinates = (10, 20, 30)
print(coordinates) 
print(coordinates[1]) 

# Set example
unique_numbers = {1, 2, 3, 4}
print(unique_numbers)  
unique_numbers.add(5)
print(unique_numbers)  
unique_numbers.remove(3)
print(unique_numbers)  

# Dictionary example
student_info = {'name': 'Areeba', 'age': 21, 'major': 'Computer Science'}
print(student_info) 
print(student_info['name']) 
student_info['age'] = 22
print(student_info) 
student_info['grade'] = 'A'
print(student_info) 
del student_info['major']
print(student_info) 
```
