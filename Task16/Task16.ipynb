# Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Plots
### 1. Create a Simple Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### 2. Plot Multiple Lines on the Same Graph

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Multiple Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

### 3. Add Markers to a Line Plot

```python
plt.figure()
plt.plot(x, y, marker='o')
plt.title('Line Plot with Markers')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### 4. Customize Line Styles and Colors

```python
plt.figure()
plt.plot(x, y, linestyle='--', color='r', marker='o')
plt.title('Customized Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Scatter Plots
### 1. Create a Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure()
plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### 2. Customize Marker Styles, Sizes, and Colors

```python
plt.figure()
plt.scatter(x, y, marker='x', s=100, c='red')
plt.title('Customized Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### 3. Plot Multiple Scatter Plots on the Same Graph

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.figure()
plt.scatter(x1, y1, color='blue', label='Dataset 1')
plt.scatter(x2, y2, color='green', label='Dataset 2')
plt.title('Multiple Scatter Plots')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

### 4. Add Labels and Titles to the Scatter Plot

```python
plt.figure()
plt.scatter(x, y)
plt.title('Scatter Plot with Labels')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Bar Charts
### 1. Create a Vertical Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

plt.figure()
plt.bar(categories, values)
plt.title('Vertical Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### 2. Create a Horizontal Bar Chart

```python
plt.figure()
plt.barh(categories, values)
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

### 3. Customize Bar Colors and Add Edge Colors

```python
plt.figure()
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Customized Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## Pie Charts
### 1. Create a Basic Pie Chart

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.figure()
plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

### 2. Customize Slice Colors and Add a Legend

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.figure()
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart with Custom Colors')
plt.legend(labels)
plt.show()
```

### 3. Explode a Slice of the Pie Chart

```python
explode = (0, 0.1, 0, 0)  # Only "explode" the 2nd slice (i.e. 'B')

plt.figure()
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Exploded Pie Chart')
plt.show()
```

### 4. Add Percentage Labels to Slices

```python
plt.figure()
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.show()
```
