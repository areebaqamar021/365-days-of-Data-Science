# Advanced plotting techniques (subplots, histograms, box plots)


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Subplots:


```python
# Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).
# Customize the subplots with titles, axis labels, and legends.

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
data = np.random.randn(1000)

# Creating the figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y1, label='sin(x)')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('x-axis')
axs[0, 0].set_ylabel('y-axis')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, y2, color='r', label='cos(x)')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('x-axis')
axs[0, 1].set_ylabel('y-axis')
axs[0, 1].legend()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]
axs[1, 0].bar(categories, values, color='b', edgecolor='black')
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')

# Histogram
axs[1, 1].hist(data, bins=20, color='g', alpha=0.7, edgecolor='black')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust the layout to prevent overlapping
plt.tight_layout()
plt.show()
```

## Histograms

```python
# Create a histogram for a given numerical data set. Customize the number of bins.
data = np.random.randn(1000)

plt.figure()
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram with 30 Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot multiple histograms on the same figure to compare different data sets.
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2

plt.figure()
plt.hist(data1, bins=30, alpha=0.5, label='Data1')
plt.hist(data2, bins=30, alpha=0.5, label='Data2')
plt.title('Multiple Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Normalize the histograms and add appropriate labels and titles.
plt.figure()
plt.hist(data1, bins=30, alpha=0.5, density=True, label='Data1')
plt.hist(data2, bins=30, alpha=0.5, density=True, label='Data2')
plt.title('Normalized Histograms')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Box Plots

```python
# Create a box plot for a given numerical data set.
data = np.random.randn(100)

plt.figure()
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```

```python
# Generate multiple box plots in a single figure to compare different data sets.
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 2
data3 = np.random.randn(100) - 2

plt.figure()
plt.boxplot([data1, data2, data3], labels=['Data1', 'Data2', 'Data3'])
plt.title('Multiple Box Plots')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```

```python
# Customize the box plots with titles, labels, and change the colors.
plt.figure()
boxprops = dict(color="blue", linewidth=2)
medianprops = dict(color="red", linewidth=2)

plt.boxplot([data1, data2, data3], labels=['Data1', 'Data2', 'Data3'], boxprops=boxprops, medianprops=medianprops)
plt.title('Customized Box Plots')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```
