## Introduction to Seaborn (visualization with Seaborn)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Datasets

```python
# Medical Cost Personal Dataset
insurance_data = pd.read_csv('insurance.csv')

# Netflix Movies and TV Shows Dataset
netflix_data = pd.read_csv('netflix_titles.csv')
```

## Scatter Plot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=insurance_data, hue='smoker')
plt.title('Medical Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Medical Charges')
plt.show()
```

## Line Plot

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='charges', hue='smoker', data=insurance_data)
plt.title('Medical Charges vs Age for Smokers and Non-Smokers')
plt.xlabel('Age')
plt.ylabel('Medical Charges')
plt.show()
```

## Bar Plot

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='release_year', y='type', data=netflix_data, estimator=lambda x: len(x) / len(netflix_data) * 100)
plt.title('Percentage of Movies and TV Shows by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.show()
```

## Count Plot

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='region', data=insurance_data)
plt.title('Count of Insurance Records by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='charges', data=insurance_data)
plt.title('Box Plot of Charges by Region')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```
