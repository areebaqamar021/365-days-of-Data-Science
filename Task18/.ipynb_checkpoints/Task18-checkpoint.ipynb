## Introduction to Seaborn (visualization with Seaborn)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Datasets

```python
# Medical Cost Personal Dataset
insurance_data = pd.read_csv('insurance.csv')

# Pokemon Dataset
pokemon_data = pd.read_csv('pokemon.csv')

# Netflix Movies and TV Shows Dataset
netflix_data = pd.read_csv('netflix_titles.csv')
```

## Scatter Plot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=insurance_data, hue='smoker')
plt.title('Medical Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Medical Charges')
plt.show()
```

## Line Plot

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Total', y='HP', data=pokemon_data)
plt.title('Pokemon Total Stats vs HP')
plt.xlabel('Total Stats')
plt.ylabel('HP')
plt.show()
```
